```python
def get_empty_coordinates(matrix):
    # return torch.nonzero(matrix == 0)
    return np.argwhere(matrix==0)

@njit
def get_empty_coordinates_nb(matrix):
    # return torch.nonzero(matrix == 0)
    return np.argwhere(matrix==0)

def get_empty_coordinates_loop(matrix):
    result = []
    h, w = len(matrix), len(matrix[0])
    # h, w = matrix.shape
    for row_index in range(h):
        for col_index in range(w):
            if matrix[row_index][col_index] == 0:
                result.append((row_index, col_index))
    return result

@njit
def get_empty_coordinates_loop_nb(matrix):
    result = []
    h, w = len(matrix), len(matrix[0])
    # h, w = matrix.shape
    for row_index in range(h):
        for col_index in range(w):
            if matrix[row_index][col_index] == 0:
                result.append((row_index, col_index))
    return result
```
